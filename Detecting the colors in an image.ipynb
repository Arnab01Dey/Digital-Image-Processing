```
# Importing required modules
import numpy as np
import cv2
import matplotlib.pyplot as plt
# Defining and reading an Image usinig opencv module
img = cv2.imread('image.png',cv2.IMREAD_COLOR)
# Convert GBR colour mode to RGB colour mode
RGBimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#Convert RGB colour mode to HSV colour mode
HSVimg = cv2.cvtColor(RGBimg,cv2.COLOR_RGB2HSV)
lower = np.array([25,150,50])
upper = np.array([35,255,255])
# To detetic a specific color eg: yellow
mask = cv2.inRange(HSVimg,lower,upper)
# To blacken out the other colors
res=cv2.bitwise_and(RGBimg, RGBimg, mask=mask)
plt.figure(figsize=(20,8))
plt.imshow(res)
```
